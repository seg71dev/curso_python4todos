## Definicion de una funcin

La sintaxis utilizada para definir funciones en Python es la siguiente:

```
def <nombre_funcion>([<parmetros>]):
    <sentencia(s)>
```

El ltimo elemento, `<sentencia(s)>`, se denomina cuerpo de la funcin. El cuerpo es el bloque de sentencias en Python que se ejecutar cuando se llame a la funcin. El cuerpo de una funcin de Python se define por la sangra.

````python
def mi_funcion(arg1, arg2):
    print(arg1)
    print(arg2)
````

### Invocar una funcin
1. La sintaxis utilizada para invocar una funcin es la siguiente: `<nombre_funcion>([<argumentos>])`
2. Los `<argumentos>` son los valores que se pasan a la funcin. Se corresponden con los `<parmetros>` en la definicin de la funcin en Python. 

````python
mi_funcion("Hola mundo", "Adios mundo")
````

````python
print(1)
mi_funcion(2, 3)
print(4)
````

## Sentencia return

Las funciones en Python pueden retornar un valor despues de haber ejecutado las sentencias de cdigo definidas. Para devolver un valor, se debe utilizar la palabra `return` dentro del cuerpo de la funcin.

Debemos de tener en cuenta varias cosas cuando utilizamos la palabra `return`:

1. **Cuando se ejecuta esta sentencia en Python, el intrprete termina inmediatamente** la ejecucin de la funcin y regresa a la lnea de cdigo desde dnde la hubisemos invocado.

````python
def mi_funcion():
    print("Sentencia 1 dentro de mi_funcion")
    return
    print("Sentencia2 dentro de mi_funcion")
    
````

````python
print("Sentencia 1 fuera de mi funcion")
mi_funcion()
print("Sentencia 2 fuera de mi funcion")
````

2. Si indicamos definimos la palabra `return` de la siguiente forma `return <valor>` se retornar el `<valor>` especificado en la llamada a dicha funcin.

````python
def mi_funcion2():
    return "Valor de mi funcion2"
    print("Esta cadena nunca se ejecutara")
````

````python
var = mi_funcion2()
print(var)
````

````python
mi_funcion2()
````

````python
mi_funcion2()[0:5]
````

````python
var = mi_funcion2()
print(var[0:5])
````

````python
print(mi_funcion2()[0:5])
````

### 5. Docstrings

Cuando la primera declaracin en el cuerpo de una funcin en Python es un string de la forma `"""texto"""`, se conoce como docstring de la funcin. 

Los docstring se utilizan para proporcionar la documentacin de una funcin. Puede contener el propsito de la funcin, los argumentos que toma, informacin sobre los valores de retorno, o cualquier otra informacin que se considere til.

Existe un PEP (Python Enhancement Proporsal) que establece el estilo con el que deben definirse los docstrings: https://www.python.org/dev/peps/pep-0257/

````python
def mi_funcion():
    """Esto es un docstring de una sola linea"""
    print("Hola Mundo")
    
````

````python
# Se ve el doc string
mi_funcion()
````

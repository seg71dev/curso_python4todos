## D. Ejercicios añadidos (propuestos) — para una base sólida

A continuación se incluyen ejercicios adicionales, creados para cubrir huecos prácticos y casos límite: están pensados para practicar validación, manejo de errores, precedencia y mini-retos integradores.

1. **Calculadora robusta de salario**

   Pide horas y tarifa; si la entrada no es numérica vuelve a pedirla hasta recibir un número válido. Imprime el salario con dos decimales.

2. **Edad y año para 100**

   Pide la edad al usuario; valida que sea un entero positivo. Imprime en qué año cumplirá 100 años (usa el año actual escrito como constante en el enunciado).

3. **Conversor avanzado Celsius↔Fahrenheit**

   Pide temperatura y escala (C o F). Convierte a la otra escala y muestra resultado formateado con 1 decimal. Controla entradas inválidas.

4. **Formateo de número favorito**

   Pide al usuario su número favorito (puede ser entero o decimal). Muestra: el número redondeado a 0, 1 y 2 decimales, y el número en notación científica.

5. **Índices seguros**

   Pide una cadena y un índice; si el índice no es válido muestra un mensaje y pide otro índice. Cuando sea válido, imprime: carácter en ese índice, su índice negativo equivalente y la subcadena desde ese índice hasta el final.

6. **Slicing avanzado**

   Pide una cadena y devuelve: cada segundo carácter, la cadena invertida sin el primer y último carácter, y la longitud original y de cada nueva cadena.

7. **Normalizador de nombres**

   Pide nombre completo con espacios extras y mayúsculas aleatorias. Devuelve nombre en formato "Apellido, Nombre" (suponiendo que el usuario introdujo al menos dos palabras), limpio y con primera letra en mayúscula por palabra. Controla entradas con una sola palabra.

8. **Contador de vocales y consonantes**

   Pide una frase y cuenta cuántas vocales y consonantes tiene (ignora espacios y signos de puntuación). Muestra ambos totales.

9. **Reemplazo condicional**

   Pide una cadena y dos caracteres `a` y `b`. Reemplaza todas las apariciones de `a` por `b`, pero si `a == b` informa que no hay cambio. Controla caso en minúsculas/mayúsculas (ofrece la opción al usuario).

10. **Mini-test con puntuación**

   Crea un cuestionario de 4 preguntas (respuestas cortas). Lee respuestas del usuario, cuenta aciertos y muestra porcentaje. Valida entradas vacías.

11. **Expresiones y precedencia (reto)**

   Escribe 8 expresiones aritméticas distintas que den el resultado 42 — usa paréntesis, potencias, divisiones enteras y flotantes. Incluye al menos una con `//` y otra con `**`.

12. **Mini-proyecto: ticket de compras**

   Pide nombre del artículo, cantidad y precio unitario (repite hasta que el usuario escriba "fin"). Al terminar, muestra lista de artículos y el total con dos decimales. Controla entradas no numéricas y cantidades negativas.
